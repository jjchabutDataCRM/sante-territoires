# Chargement des APLs 

```python
import pandas as pd
from pathlib import Path

# On essaie d'utiliser __file__, sinon on prend le dossier actuel
try:
    BASE_DIR = Path(__file__).resolve().parent
except NameError:
    BASE_DIR = Path.cwd()

ROOT = BASE_DIR.parent
DATA_DIR = ROOT / "data/raw/offres_soins"

%ls ../data/raw/offres_soins

print(DATA_DIR)


dfs_2022 = []
```

## 1  - Chargement APL médecins

```python
print("1 - Chargement APL Médecins généralistes ...")
file_path = DATA_DIR / 'apl_medecins_generalistes.xlsx'

df_apl_medecins = pd.read_excel(
    #'../data/raw/offres_soins/apl_medecins_generalistes.xlsx',
    file_path,
    sheet_name='APL 2023',
    skiprows=8,
    dtype={'Code commune INSEE': str}
).rename(columns={
    'Code commune INSEE': 'code_insee_comm',
    'Commune': 'nom_commune',
    'APL aux médecins généralistes': 'apl_medecins',
    'APL aux médecins généralistes de 65 ans et moins ': 'apl_65ans',
    'APL aux médecins généralistes de 62 ans et moins ': 'apl_62ans',
    'APL aux médecins généralistes de 60 ans et moins ': 'apl_60ans',
    'Population standardisée 2021 pour la médecine générale': 'pop_standard_2021',
    'Population totale 2021': 'pop_totale_2021'
    })

print(f"{len(df_apl_medecins):,} lignes chargées")
print(f"Colonnes : {df_apl_medecins.columns.tolist()}")

df_apl_med_2022 = pd.read_excel( 
        file_path,
        sheet_name='APL 2022',
        skiprows=8, dtype={'Code commune INSEE': str}
    ).rename(columns={
        'Code commune INSEE': 'code_insee_comm',
        'APL aux médecins généralistes': 'apl_medecins_2022',
        'APL aux médecins généralistes de 65 ans et moins ': 'apl_65a_2022',
        'APL aux médecins généralistes de 62 ans et moins ': 'apl_62a_2022',
        'APL aux médecins généralistes de 60 ans et moins ': 'apl_60a_2022',}
        )
[['Code commune INSEE', 'apl_medecins_prec', 'apl_65a_prec', 'apl_62a_prec', 'apl_60a_prec']]

dfs_2022.append(df_apl_med_2022)
```

## 2  - Chargement APL dentistes

```python
print('-'*50)
print("2 - Chargement APL chirurgiens-dentistes ...")
file_path = DATA_DIR / 'apl_chirurgiens_dentistes.xlsx'
df_apl_dentistes = pd.read_excel(
    file_path,
    sheet_name='APL 2023',
    skiprows=8,
    dtype={'Code commune INSEE': str}
).rename(columns={'APL aux chirurgiens-dentistes': 'apl_dentistes',
                  'Code commune INSEE': 'code_insee_comm',
                  'Commune': 'nom_commune'})

print(f"{len(df_apl_dentistes):,} lignes chargées")
print(f"Colonnes : {df_apl_dentistes.columns.tolist()}")
print("\n🔍 Vérification des valeurs :")
print(df_apl_dentistes[['code_insee_comm', 'nom_commune', 'apl_dentistes']].head(10))

df_apl_dent_2022 = (
    pd.read_excel(
        file_path,
        sheet_name='APL 2022',
        skiprows=8,
        dtype={'Code commune INSEE': str}
    )
    .rename(columns={'APL aux chirurgiens-dentistes': 'apl_dentistes_2022',
                     'Code commune INSEE': 'code_insee_comm'})
    [['code_insee_comm', 'apl_dentistes_2022']]
) 
dfs_2022.append(df_apl_dent_2022)
```

## 3  - Chargement APL Infirmières

```python
# 3 Infirmiéres
print('-'*50)
print("3 - Chargement APL Infirmières ...")
file_path = DATA_DIR / 'apl_infirmieres.xlsx'
df_apl_infirmiers = pd.read_excel(
    file_path,
    sheet_name='APL 2023',
    skiprows=8,
    dtype={'Code commune INSEE': str}
).rename(columns={'APL aux infirmières': 'apl_infirmiers',
                  'Code commune INSEE': 'code_insee_comm',
                  'Commune': 'nom_commune'}) 

print(f"{len(df_apl_infirmiers):,} lignes chargées")
print(f"Colonnes : {df_apl_infirmiers.columns.tolist()}")
print("\n🔍 Vérification des valeurs :")
print(df_apl_infirmiers[['code_insee_comm', 'nom_commune', 'apl_infirmiers']].head(10))


df_apl_infirmiers_2022 = (
    pd.read_excel(
        file_path,
        sheet_name='APL 2022',
        skiprows=8,
        dtype={'Code commune INSEE': str}
    )
    .rename(columns={'APL aux infirmières': 'apl_infirmiers_2022',
                     'Code commune INSEE': 'code_insee_comm'})
    [['code_insee_comm', 'apl_infirmiers_2022']]
) 
dfs_2022.append(df_apl_infirmiers_2022)
```

## 4  - Chargement APL kinés

```python
# 4. Kinés
print('-'*50)
print("4 - Chargement APL kinésithérapeutes ...")
file_path = DATA_DIR / 'apl_kinesitherapeutes.xlsx'
df_apl_kines = pd.read_excel(
    file_path,
    sheet_name='APL 2023',
    skiprows=8,
    dtype={'Code commune INSEE': str}
).rename(columns={'APL aux kinésithérapeutes': 'apl_kines',
                  'Code commune INSEE': 'code_insee_comm',
                  'Commune': 'nom_commune'}) 

print(f"{len(df_apl_kines):,} lignes chargées")
print(f"Colonnes : {df_apl_kines.columns.tolist()}")

df_apl_kines_2022 = (
    pd.read_excel(
        file_path,
        sheet_name='APL 2022',
        skiprows=8,
        dtype={'Code commune INSEE': str}
    )
    .rename(columns={'APL aux kinésithérapeutes': 'apl_kines_2022',
                     'Code commune INSEE': 'code_insee_comm'})
    [['code_insee_comm', 'apl_kines_2022']]
) 
dfs_2022.append(df_apl_kines_2022)
```

## 5  - Chargement APL sages-femmes

```python
# 5. Sages-femmes
print('-'*50)
print("5 - Chargement APL sages-femmes ...")
file_path = DATA_DIR / 'apl_sages_femmes.xlsx'
df_apl_sagesfemmes = pd.read_excel(
    file_path,
    sheet_name='APL 2023',
    skiprows=8,
    dtype={'Code commune INSEE': str}
).rename(columns={'APL aux sages-femmes': 'apl_sagesfemmes',
                  'Code commune INSEE': 'code_insee_comm',
                  'Commune': 'nom_commune'
                  })

print(f"{len(df_apl_sagesfemmes):,} lignes chargées")
print(f"Colonnes : {df_apl_sagesfemmes.columns.tolist()}")

df_apl_sagesfemmes_2022 = (
    pd.read_excel(
        file_path,
        sheet_name='APL 2022',
        skiprows=8,
        dtype={'Code commune INSEE': str}
    )
    .rename(columns={'APL aux sages-femmes': 'apl_sagesfemmes_2022',
                     'Code commune INSEE': 'code_insee_comm'})
    [['code_insee_comm', 'apl_sagesfemmes_2022']]
) 
dfs_2022.append(df_apl_sagesfemmes_2022)
```

```python
for i, df in enumerate(dfs_2022):
    if "code_insee_comm" not in df.columns:
        print(f"⚠️ Erreur au DataFrame index {i} !")
        print(f"Colonnes réelles : {df.columns.tolist()}")
```

```python
from functools import reduce

df_apls_2022 = reduce(
    lambda left, right: pd.merge(left, right, on="code_insee_comm", how="outer"),
    dfs_2022
)
```

```python


df_merged = df_apl_medecins\
    .merge(df_apl_dentistes[['code_insee_comm', 'apl_dentistes']], 
           left_on='code_insee_comm', right_on='code_insee_comm', how='left', suffixes=('', '_y'))\
    .merge(df_apl_infirmiers[['code_insee_comm', 'apl_infirmiers']], 
           left_on='code_insee_comm', right_on='code_insee_comm', how='left', suffixes=('', '_y'))\
    .merge(df_apl_kines[['code_insee_comm', 'apl_kines']], 
           left_on='code_insee_comm', right_on='code_insee_comm', how='left', suffixes=('', '_y'))\
    .merge(df_apl_sagesfemmes[['code_insee_comm', 'apl_sagesfemmes']], 
           left_on='code_insee_comm', right_on='code_insee_comm', how='left', suffixes=('', '_y'))


print(f"✅ {len(df_merged):,} communes avec apls après jointure")
print(f"Colonnes APL : {[c for c in df_merged.columns if 'apl_' in c]}")
print(df_merged.columns)
```

```python
# Ajout des apls 2022
df_final["apl_medecins_2022"] = df_apls_2022["apl_medecins_2022"]
df_final["apl_dentistes_2022"] = df_apls_2022["apl_dentistes_2022"]
df_final["apl_infirmiers_2022"] = df_apls_2022["apl_infirmiers_2022"]
df_final["apl_kines_2022"] = df_apls_2022["apl_kines_2022"]
df_final["apl_sagesfemmes_2022"] = df_apls_2022["apl_sagesfemmes_2022"]
```

```python
# supprimer la deuxiéme ligne d'entête
df_final.dropna(inplace=True)
```

```python
# Calcul des deltas
df_final["delta_apl_medecins"] = df_final["apl_medecins"] - df_final["apl_medecins_2022"]
df_final["delta_apl_dentistes"] = df_final["apl_dentistes"] - df_final["apl_dentistes_2022"]
df_final["delta_apl_infirmiers"] = df_final["apl_infirmiers"] - df_final["apl_infirmiers_2022"]
df_final["delta_apl_kines"] = df_final["apl_kines"] - df_final["apl_kines_2022"]
df_final["delta_apl_sagesfemmes"] = df_final["apl_sagesfemmes"] - df_final["apl_sagesfemmes_2022"]
```

```python
# Supprimer les colonnes 2022
cols_2022 = [ "apl_medecins_2022", "apl_dentistes_2022", "apl_infirmiers_2022", "apl_kines_2022", "apl_sagesfemmes_2022" ]
```

```python
df_final.columns
```

```python
# Standardiser les APLs
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Ajouter une colonne pour chaque APL standardisée 'NomColonne'_std
for col in ['apl_medecins', 'apl_dentistes', 'apl_infirmiers', 'apl_kines', 'apl_sagesfemmes']:
    df_final[col + '_std'] = scaler.fit_transform(df_final[[col]])
```

```python
weights = {
    'apl_med_consult': 0.35,
    'apl_inf_etp': 0.25,
    'apl_kine_etp': 0.20,
    'apl_dent_etp': 0.15,
    'apl_sf_etp': 0.05
}
```

```python
df_final['score_apl'] = (
    df_final['apl_medecins_std'] * weights['apl_med_consult'] +
    df_final['apl_infirmiers_std'] * weights['apl_inf_etp'] +
    df_final['apl_kines_std'] * weights['apl_kine_etp'] +
    df_final['apl_dentistes_std'] * weights['apl_dent_etp'] +
    df_final['apl_sagesfemmes_std'] * weights['apl_sf_etp']
)
```

```python
# Catégorisation en quintiles
df_final['quintile_apl_nat'] = pd.qcut(df_final['score_apl'], q=5, labels=['Q1 (très faible)', 'Q2 (faible)', 'Q3 (moyen)', 'Q4 (bon)', 'Q5 (très bon)'])
```
