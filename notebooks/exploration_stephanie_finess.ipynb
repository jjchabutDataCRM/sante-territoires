```python
import pandas as pd
import json

csv_path = '../data/raw/reexposition_finess/finess_equipements_sociaux.txt'
schema_path = 'votre_fichier_schema_equipements_sociaux.json'

# Lire CSV
df = pd.read_csv(csv_path, nrows=1, sep=";", encoding="utf-8")

# Colonnes à forcer en STRING
cols_fix = [c for c in df.columns if 'code_' in c.lower() or 'siren' in c.lower()]

# Générer schéma
schema = []
for col in df.columns:
    bq_type = "STRING" if col in cols_fix or df[col].dtype == 'object' else \
              "INTEGER" if df[col].dtype in ['int64', 'int32'] else \
              "FLOAT" if df[col].dtype in ['float64', 'float32'] else "STRING"
    schema.append({"name": col, "type": bq_type, "mode": "NULLABLE"})

# Sauvegarder
with open(schema_path, 'w') as f:
    json.dump(schema, f, indent=2)

print(f"✅ {schema_path}")
```
