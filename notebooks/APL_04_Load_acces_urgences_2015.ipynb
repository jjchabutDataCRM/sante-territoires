# Chargement des données d'accès aux urgences 2015

```python
import pandas as pd
from pathlib import Path

# On essaie d'utiliser __file__, sinon on prend le dossier actuel
try:
    BASE_DIR = Path(__file__).resolve().parent
except NameError:
    BASE_DIR = Path.cwd()

ROOT = BASE_DIR.parent
DATA_DIR = ROOT / "data"

%ls ../data/raw/offres_soins

print(DATA_DIR)
```

```python
file_path = DATA_DIR / "raw/offres_soins/Diagnostic d'accès aux soins urgents au 31-12-2015.xls"

# Voir les onglets disponibles
xl = pd.ExcelFile(file_path)
print(f"Onglets disponibles : {xl.sheet_names}")
```

```python
# Charger l'onglet BASECOM_URGENCES_2015
df = pd.read_excel(
    file_path,
    sheet_name='BASECOM_URGENCES_2015',
    skiprows=5,
    #dtype={'Code commune': str}  # Garder les zéros initiaux
)
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
file_path = DATA_DIR / 'extracted' / 'raw_comm_acces_urgences.csv'
df.to_csv(file_path)

file_path = DATA_DIR / 'extracted' / 'raw_comm_acces_urgences.parquet'
df.to_parquet(file_path)
```
