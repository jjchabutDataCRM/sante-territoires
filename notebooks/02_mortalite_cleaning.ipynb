# 1. Nettoyage taux brut de mortalité

```python
import pandas as pd
from pathlib import Path

raw_path = Path("../data/raw")
processed_path = Path("../data/processed")

processed_path.mkdir(exist_ok=True)
```

```python
print("Raw exists:", raw_path.exists())
print("Processed exists:", processed_path.exists())
```

```python
df_brut = pd.read_excel(raw_path / "mortalite_2023_taux_brut.xlsx")
df_brut.head()
```

```python
xls = pd.ExcelFile(raw_path / "mortalite_2023_taux_brut.xlsx")
xls.sheet_names
```

```python
df_brut = pd.read_excel(raw_path / "mortalite_2023_taux_brut.xlsx", sheet_name="Données")
df_brut.head()
```

```python
df_brut.info()
```

```python
df_brut.columns = [
    "annee",
    "departement",
    "sexe",
    "cause",
    "indicateur",
    "valeur"
]
df_brut.head()
```

```python
df_brut.groupby(["departement", "sexe"]).size()
```

```python
df_brut.duplicated().sum()
```

```python
df_brut["valeur"].describe()
```

```python
df_brut.groupby("cause")["valeur"].mean().sort_values(ascending=False)
```

```python
df_brut.groupby("cause")["valeur"].describe()
```

```python
df_brut.groupby("cause")["valeur"].std().sort_values(ascending=False)
```

```python
df_brut.to_csv(
    processed_path / "mortalite_2023_taux_brut_clean.csv",
    index=False
)
```

# Nettoyage effectif mortalité (df_eff)

```python
df_eff = pd.read_excel(raw_path / "mortalite_2023_effectif.xlsx", sheet_name="Données")
df_eff.head()
```

```python
df_eff.columns = [
    "annee",
    "departement",
    "sexe",
    "cause",
    "indicateur",
    "valeur"
]
```

```python
set(df_brut.columns) == set(df_eff.columns)
```

# Concatenation des deux dataframes df_brut et df_eff pour taux brut et effectif de de la mortalité en 2023 en Occitanie dans un seul dataframe df_all

```python
df_all = pd.concat([df_brut, df_eff], ignore_index=True)
df_all["indicateur"].value_counts()
```

```python
df_eff.info()
```

```python
df_eff.describe()
```

```python
df_eff.head()
```

```python
df_all.head()
```

```python
df_all.to_csv(
    processed_path / "mortalite_2023_observee_clean.csv",
    index=False
)
```
