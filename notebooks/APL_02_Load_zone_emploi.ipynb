```python
import geopandas as gpd

df_georef_ze = gpd.read_file('../data/raw/territoires/georef-france-zone-emploi-2020.geojson')
```

```python
df_georef_ze.head()
```

```python
df_georef_ze.shape
```

```python
df_georef_ze.columns
```

```python
df_ze_final = df_georef_ze.copy()
df_ze_final = df_ze_final.drop(columns=['year','ze2020_current_code', 'ze2020_name_upper', 'ze2020_name_lower', 'ze2020_area_code', 'ze2020_type'])
df_ze_final.head()
```

```python
if "geo_point_2d" in df_ze_final.columns:
    df_ze_final["lon"] = df_ze_final["geo_point_2d"].apply(lambda d: d.get("lon") if isinstance(d, dict) else None) 
    df_ze_final["lat"] = df_ze_final["geo_point_2d"].apply(lambda d: d.get("lat") if isinstance(d, dict) else None)
```

```python
for col in ["reg_code", "reg_name", "dep_code", "dep_name", "ze2020_code", "ze2020_name"]: 
    df_ze_final[col] = df_ze_final[col].apply(lambda x: x[0] if isinstance(x, list) and len(x) == 1 else x)

df_ze_final.head()
```

```python
import numpy as np

def unwrap_array(v):
    # Cas 1 : numpy array
    if isinstance(v, np.ndarray):
        if len(v) == 1:
            return v[0]              # ex: array(['32']) → '32'
        else:
            return ",".join(map(str, v))  # ex: array(['46','82']) → '46,82'
    return v

cols = ["reg_code", "reg_name", "dep_code", "dep_name", "ze2020_code", "ze2020_name"]

for col in cols:
    df_ze_final[col] = df_ze_final[col].apply(unwrap_array)
```

```python
df_ze_final
```

```python
df_geom_ze = df_ze_final.copy()

df_geom_ze = df_geom_ze[['ze2020_code', 'ze2020_name', 'geometry', 'lon', 'lat']]

df_geom_ze.head()
```

```python
df_geom_ze.crs
```

```python
def simplify_geom(gdf, tolerance=20):
    # Reprojection en Lambert 93 car pandas peut sans qu'on le sache le charger en WGS84 
    # la simplification est plus efficace dans un système métrique
    gdf = gdf.to_crs(2154)

    # Correction + simplification + correction
    gdf["geometry"] = (
        gdf["geometry"]
        .buffer(0)
        .simplify(tolerance, preserve_topology=False)
        .buffer(0)
    )

    # Reprojection finale en WGS84 pour le web
    return gdf.to_crs(4326)
```

```python
df_geom_ze = simplify_geom(df_geom_ze, 100)
```

```python
df_geom_ze.to_csv('../data/processed/geom/zone_emploi_geom_simplified.csv', index=False)
df_geom_ze.to_parquet('../data/processed/geom/zone_emploi_geom_simplified.parquet', index=False)
```
